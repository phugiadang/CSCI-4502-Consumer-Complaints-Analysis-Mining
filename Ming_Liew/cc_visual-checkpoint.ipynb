```python
from collections import Counter

#rank all attributes
def ranking(attr_a,):
    loc_col_company,loc_col_freq = [],[]
    loc_attr_a = [i for i in attr_a if i != ""]
    freq = Counter(loc_attr_a)
    for n,v in freq.most_common(100):
        loc_col_company.append(n),loc_col_freq.append(v)
    return (freq.most_common(100),loc_col_company,loc_col_freq)
(freq,col_freq,col_company) = ranking(company)
for i in freq:
    print i,'\n'
```
